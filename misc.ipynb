```python
import pandas as pd
```

```python
candidatos = pd.read_excel('./data/interim/candidatos_2018_minimo.xlsx')
```

```python
(
    candidatos
    .loc[lambda df: df.NR_TURNO == 1]
    .groupby([
        'DS_COR_RACA',
        'SG_PARTIDO',
    ])
    .size()
    .rename('TOTAL')
    .reset_index()
    .sort_values(['DS_COR_RACA', 'TOTAL'], ascending=(True, False))
    .set_index(['DS_COR_RACA', 'SG_PARTIDO'])
)
```

```python
with pd.option_context('display.max_rows', None):
    display(
        candidatos
        .groupby([
            'SG_PARTIDO',
            'DS_COR_RACA',
        ])
        .size()
        .rename('TOTAL')
        .reset_index(['DS_COR_RACA'])
        .join(
            candidatos
            .groupby('SG_PARTIDO')
            .size()
            .rename('TOTAL_PARTIDO')
        )
        .assign(
            PROPORCAO=lambda df: df.TOTAL / df.TOTAL_PARTIDO
        )
        .reset_index()
        .loc[lambda df: df.groupby('SG_PARTIDO').PROPORCAO.idxmax()]
        .set_index('SG_PARTIDO')
    )
```
