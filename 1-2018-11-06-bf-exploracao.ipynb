## O que são os suplentes?

## Quantos partidos participaram da eleição

```python
len(candidatos.SG_PARTIDO.unique())
```

## Quantos partidos por turno?

```python
candidatos.groupby(['NR_TURNO']).SG_PARTIDO.nunique()
```

## Quantos candidatos por turno?

```python
candidatos.groupby([
    'NR_TURNO'
]).size()
```

## Quantos candidatos por turno, por partido?

```python
_ = candidatos.groupby([
    'NR_TURNO',
    'SG_PARTIDO'
]).size().rename('Total Candidatos')

(
    _
    .to_frame()
    .reset_index()
    .sort_values(['NR_TURNO', 'Total Candidatos'], ascending=[True, False])
    .set_index(['NR_TURNO', 'SG_PARTIDO'])
)
```

## Quantos candidatos por turno, por UF

```python
candidatos.groupby([
    'NR_TURNO',
    'SG_UF',
]).size()
```

## Porque os estados com segundo turno tem todos 4 candidatos?

```python
_[_.SG_UF == 'SP']
```

#### Porque esses cargos são de ambos os candidatos a governadores e seus vices

#### Nota especial para UF == BR:

```python
_ = candidatos[candidatos.NR_TURNO == 2].copy()
_[_.SG_UF == 'BR']
```

## Quantos candidatos por turno, por cargo

```python
(
    candidatos
    .groupby([
        'NR_TURNO',
        'DS_CARGO',
    ])
    .size()
    .rename('Total Candidatos')
    .reset_index()
    .sort_values(['NR_TURNO', 'Total Candidatos'], ascending=[True, False])
    .set_index(['NR_TURNO', 'DS_CARGO'])
)
```

## Quantos candidatos por cargo por raça

```python
(
    candidatos
    [candidatos.NR_TURNO == 1]
    .groupby([
        'DS_CARGO',
        'DS_COR_RACA',
    ])
    .size()
    .rename('Total Candidatos')
    .reset_index()
    .sort_values(['DS_CARGO', 'Total Candidatos'], ascending=[True, False])
    .set_index(['DS_CARGO', 'DS_COR_RACA'])
)
```
